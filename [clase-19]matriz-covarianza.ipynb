```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
import plotly.express as px

iris = sns.load_dataset('iris')
```

```python
g = sns.pairplot(iris, 
                 hue = 'species', 
                 palette='Set2')

g.fig.set_size_inches(16,8)


g.fig.suptitle('Graficas de correlacion de dataset iris')

plt.show()
```

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(
    iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
)
scaled.T
```

```python
covariance_matrix = np.cov(scaled.T)
covariance_matrix
```

```python
plt.figure(figsize=(10,10))
sns.set(font_scale=1.5)
hm = sns.heatmap(covariance_matrix,
                 cbar=True,
                 annot=True,
                 square=True,
                 fmt='.2f',
                 annot_kws={'size': 12},
                 yticklabels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
                 xticklabels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
                 cmap = 'coolwarm')

plt.show()
```

```python
plt.figure(figsize=(10,10))

sns.heatmap(iris.corr(), annot=True, cmap = 'coolwarm')

plt.show()
```

```python
fig = px.imshow(iris.corr(),
                color_continuous_scale='Portland')

fig.update_layout(title = 'Heatmaps de correlacion de nuestro dataset iris',
                 font = dict(family = 'verdana', size = 16, color = 'white'),
                 template = 'plotly_dark',
                 height = 750,
                 width = 900)

fig.show()
```
