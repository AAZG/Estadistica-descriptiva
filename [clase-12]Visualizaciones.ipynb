```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go

iris = sns.load_dataset('iris')

iris.head()
```

https://seaborn.pydata.org/generated/seaborn.scatterplot.html

```python
# scatterplot 
plt.figure(figsize=(16,9))

sns.scatterplot(data=iris, 
                x='sepal_length', 
                y='petal_length', 
                hue='species', 
                palette='Set2',
                alpha = 0.8)

plt.show()
```

```python
linea = pd.DataFrame(dict(x=[0,8],
                    y=[0,8]))
```

```python
linea
```

```python
fig1 = px.scatter(iris,
                  x="sepal_length", 
                  y="petal_length", 
                  color='species', 
                  color_discrete_sequence=['rgb(175, 100, 88)', '#479B55', '#FA0087'],
                  opacity=0.8)

fig2 = px.line(linea, 
               x='x', 
               y='y')

fig2.update_traces(line_color='blue')

fig3 = go.Figure(data=fig1.data + fig2.data)

fig3.update_layout(title='Longitud de petalo/Longitud sepalo',
                   xaxis_title='Longitud de sepalo',
                   yaxis_title='Longitud de petalo',
                   font=dict(family='verdana', size = 16, color = 'white'),
                   template='plotly_dark',
                   height=500,
                   width=900,
                   legend_bgcolor='rgb(117, 112, 179)')

fig3.show()
```

http://seaborn.pydata.org/generated/seaborn.jointplot.html

```python
# joint plot

plt.figure(figsize=(16,9))

sns.jointplot(data=iris, 
              x='sepal_length', 
              y='petal_length', 
              hue='species',
              palette='Set2',
              alpha = 0.8)

plt.show()
```

```python
fig1 = px.scatter(iris, 
                  x="sepal_length", 
                  y="petal_length", 
                  color='species',
                  color_discrete_sequence=['rgb(175, 100, 88)', '#479B55', '#FA0087'],
                  marginal_x="box",
                  marginal_y="box",
                  opacity=0.8)

fig1.update_layout(title='Longitud de petalo/Longitud sepalo',
                   xaxis_title='Longitud de sepalo',
                   yaxis_title='Longitud de petalo',
                   font = dict(family='verdana', size = 16, color = 'white'),
                   template='plotly_dark',
                   height=500,
                   width=900,
                   legend_bgcolor='rgb(117, 112, 179)')

fig1.show()
```

https://seaborn.pydata.org/generated/seaborn.boxplot.html

```python
# box plot 

plt.figure(figsize=(14,7))

sns.boxplot(x='species', 
            y='sepal_length', 
            data=iris, 
            palette='Set2')

plt.show()
```

```python
fig = px.box(iris, 
             x='species', 
             y="sepal_length", 
             color='species',
             color_discrete_sequence=['rgb(175, 100, 88)', '#479B55', '#FA0087'])

fig.update_layout(title='Comparativa Tamaño sepalo flor por especie',
                  xaxis_title='Especie',
                  yaxis_title='Longitud de petalo',
                  font=dict(family='verdana', size = 16, color = 'white'),
                  template='plotly_dark',
                  height=500,
                  width=900,
                  legend_bgcolor='rgb(117, 112, 179)')

fig.show()
```

https://seaborn.pydata.org/generated/seaborn.barplot.html

```python
#Barplot

plt.figure(figsize=(14,7))

sns.barplot(x='species', 
            y='sepal_length', 
            data=iris, 
            palette='Set2',
            alpha = 0.8)

plt.show()
```

```python
fig = px.bar(iris, 
             x='species', 
             y="sepal_length", 
             color='species', 
             color_discrete_sequence=['rgb(175, 100, 88)', '#479B55', '#FA0087'], 
             barmode='overlay',
             opacity=0.7)

fig.update_layout(title='Comparativa Tamaño sepalo flor por especie',
                  xaxis_title='Especie',
                  yaxis_title='Longitud de petalo',
                  font=dict(family = 'verdana', size = 16, color = 'white'),
                  template='plotly_dark',
                  height=500,
                  width=900,
                  legend_bgcolor='rgb(117, 112, 179)', 
                  xaxis={'categoryorder':'category descending'})

fig.show()
```
